```python

dayLst = corrlab.generateDayLst(start_date='20171101', end_date='20171130', freq='B')

conn = MySQLdb.connect(host='localhost', user='root', passwd='hhui123456')
cursor = conn.cursor()
cursor.execute("""create database if not exists db_corr""")
conn.select_db('db_corr')

res = pd.DataFrame()
for day in dayLst:   # 时间跨度为1天
    lst = []
    lst.append(day)
    data = corrlab.concatdata(lst, keywd='mid_price_return_cum')
    target = getticker(data, ticker1)
    print 'processing day:',day

    for lag in ['1s','5s','10s','30s','60s']:
        for type in ['1s','5s','10s','30s','60s']:
            res = pd.DataFrame()
            temp = data.copy()
            shifted = temp[target].shift(-int(lag[:-1]), 's')
            align_base = corrlab.get_align_base(data)
            _, align_shifted = align_base.align(shifted, join='left', axis=0)
            temp[target] = align_shifted.values
            temp = corrlab.sampledata(temp, period = type)
            temp.fillna(method='ffill',inplace=True)
            temp.fillna(method = 'bfill',inplace=True)
            temp_corr = temp.corr().sort_index()
            res = pd.concat([res,temp_corr[target]])
            res.rename(columns = {0:day},inplace=True)
            res.fillna(-2,inplace=True)
            for symbol in res.index.values:
                corr_value = res[day][symbol]
                ticker2 = symbol[:2]
                cursor.execute("""REPLACE INTO tb_corr(
                            start_date,
                            end_date,
                            ticker1,
                            symbol1,
                            ticker2,
                            symbol2,
                            type,
                            lag,
                            corr)
                            VALUES (
                            '%s', '%s','%s','%s','%s','%s','%d','%d','%.3f'
                            )
                            """ % (day, day, ticker1, target, ticker2, symbol, int(type[:-1]), int(lag[:-1]), corr_value))
                conn.commit()

# 时间跨度为7天 (5个工作日)


length = len(dayLst)
for i in range(5,length):
    lst = dayLst[i-5:i]
    data = corrlab.concatdata(lst, keywd='mid_price_return_cum')
    target = getticker(data, ticker1)

    for lag in ['1s','5s','10s','30s','60s']:
        for type in ['1s','5s','10s','30s','60s']:
            res = pd.DataFrame()
            temp = data.copy()
            shifted = temp[target].shift(-int(lag[:-1]), 's')
            align_base = corrlab.get_align_base(data)
            _, align_shifted = align_base.align(shifted, join='left', axis=0)
            temp[target] = align_shifted.values
            temp = corrlab.sampledata(temp, period=type)
            temp.fillna(method='ffill', inplace=True)
            temp.fillna(method='bfill', inplace=True)
            temp_corr = temp.corr().sort_index()
            res = pd.concat([res, temp_corr[target]])
            res.rename(columns={0: lst[0]+'-'+lst[-1]}, inplace=True)
            res.fillna(-2, inplace=True)
            for symbol in res.index.values:
                corr_value = res[day][symbol]
                ticker2 = symbol[:2]
                cursor.execute("""REPLACE INTO tb_corr(
                            start_date,
                            end_date,
                            ticker1,
                            symbol1,
                            ticker2,
                            symbol2,
                            type,
                            lag,
                            corr)
                            VALUES (
                            '%s', '%s','%s','%s','%s','%s','%d','%d','%.3f'
                            )
                            """ % (lst[0], lst[-1], ticker1, target, ticker2, symbol, int(type[:-1]), int(lag[:-1]), corr_value))
                conn.commit()
    # del data, target, res; gc.collect()
```

```python
cursor.execute('select * from tb_corr where type = 1 and start_date = end_date' )
ans = cursor.fetchall()
res = {}
for i in ans:
    if i[3] in res.keys():
        if str(i[5]) in res[i[3]].keys():
            res[i[3]][str(i[5])].append(i[6])
        else:
            res[i[3]][str(i[5])] = [i[6]]
    else:
        res[i[3]] = {}
        res[i[3]][str(i[5])] = [i[6]]
        
    
import matplotlib.pylab as plt
%matplotlib inline
var = 'hc'
plt.plot(range(len(res[var]['1'])),res[var]['1'],label = var +' 1')
plt.plot(range(len(res[var]['5'])),res[var]['5'],label = var+' 5')
plt.plot(range(len(res[var]['10'])),res[var]['10'],label = var+' 10')
plt.plot(range(len(res[var]['30'])),res[var]['30'],label = var+' 30')
plt.plot(range(len(res[var]['60'])),res[var]['60'],label = var+' 60')
plt.legend(loc = 'best')
plt.show()
```
